```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])

x=np.arange(1,11)
a1.plot(x,np.exp(x),zorder=1)
a1.set_ylabel('exp')

a2=a1.twinx()
a2.plot(x,np.log(x),zorder=1)
#a2.grid(False)
#a1.grid(False)
a1.set_axisbelow(True)
a2.set_axisbelow(True)
```
